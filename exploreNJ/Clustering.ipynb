```python
from sklearn import preprocessing
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
df = pd.read_pickle('d2.pkl')
df.iloc[:,:5].info()
```

```python
X = df.iloc[:,5:]
```

```python
#X.iloc[:,44:46].describe()
#X['hh_inc_med'] = X['hh_inc_med']/1000
#X['hh_inc_mean'] = X['hh_inc_mean']/1000
#X['hu_val_median'] = X['hu_val_median']/1000

X.iloc[:,140:160].describe()
```

```python
X.drop('CYEqualizedPropertyValue(Pre-Appeal)', axis=1, level=None, inplace=True)
```

```python
X_scaled = preprocessing.scale(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(X_scaled)
labels = kmeans.labels_
kmeans.score(X_scaled)
```

```python
metrics.silhouette_score(X_scaled, labels, metric='euclidean')
```

```python
metrics.calinski_harabaz_score(X_scaled, labels)  
```

```python
metrics.silhouette_samples(X_scaled, labels, metric='euclidean')
```
