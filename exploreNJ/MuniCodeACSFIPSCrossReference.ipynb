## Extract table with muni_codes and census FIPS correlation

```python
json_data=open('nj_munis_3.json').read()
```

```python
import pandas as pd
import json
parsed_json = json.loads(json_data)

parsed_json['objects'].keys()

```

```python
parsed_json['objects']['nj_munis_2']['geometries'][0]['properties']
```

```python
base = parsed_json['objects']['nj_munis_2']['geometries']
len(base)
```

```python
muni_list = []

for i in range(565):
    info = (base[i]['properties']['census2010'],base[i]['properties']['mun_code'])
    muni_list.append(info)
```

```python
muni = pd.DataFrame(muni_list)
muni.columns = ['FIPS','Unicode']
muni.head()
```

```python
muni['FIPS'] = muni['FIPS'].astype(str)
muni.info()
```

```python
muni.to_pickle('muni.pkl')
```

```python
muni_list2 = []

for i in range(565):
    info = (base[i]['properties']['census2010'],base[i]['properties']['mun'].lower(),base[i]['properties']['county'].lower())
    muni_list2.append(info)
```

```python
muni_list2
```

```python
muni2 = pd.DataFrame(muni_list2)
muni2.columns = ['FIPS','town','county']
muni2.head()
```

```python
muni2['FIPS'] = muni2['FIPS'].astype(str)
muni2.info()
```

```python
muni2.to_pickle('muni_names.pkl')
```
