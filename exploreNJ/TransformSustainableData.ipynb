### Sustainable NewJersey transform

```python
import pandas as pd
```

```python
sustain = pd.read_csv('Data/sustainable.csv')
sustain = sustain.iloc[:,1:]
```

```python
sustain['town'] = sustain['town'].str.lower()
sustain['county'] = sustain['county'].str.lower()
```

```python
num_badge = []
for b in sustain['badge']:
    if b == 'none' : 
        num_badge.append(1)
    elif b == 'bronze':
        num_badge.append(2)
    elif b == 'silver':
        num_badge.append(3)
    else:
        num_badge.append(99)
```

```python
sustain['num_badge'] = num_badge
```

```python
sustain.describe()
sustain.info()
```

```python

muni_names = pd.read_pickle('muni_names.pkl')
muni_names.info()
```

```python
result = pd.merge(muni_names, sustain, how='left', on=['town', 'county'])
```
