```python
total_df = pd.read_pickle('total_df.pkl')
```

```python
temp = total_df.loc[200:400,:]
temp['keyword'].value_counts()
temp.info()
```

```python
tlist = temp['summary']
```

```python
from nltk.corpus import stopwords


nlist = []

rx = re.compile(r'(?<=[a-z])(?=[A-Z])')

for i in tlist:
    cln = i.replace('\n',' ')
    cln = rx.sub(' ',cln)
    #cln = [w for w in cln if w.lower() not in stopwords.words('english')]
    cln =''.join( [w for w in cln if w not in string.punctuation])
    cln = cln.lower()
    nlist.append (cln)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.feature_extraction.text import TfidfTransformer
```
