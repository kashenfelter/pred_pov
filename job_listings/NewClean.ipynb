```python
import pandas as pd
import numpy as np
import re
import nltk
import string
from nltk.corpus import stopwords
from textblob import TextBlob
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
total_df = pd.read_pickle('total_df.pkl')
```

```python
temp = total_df.loc[200:400,:]
print(temp['keyword'].value_counts())
temp.info()
```

```python
#tlist = temp['summary']
tlist = total_df['summary']

#keyword_target = temp['keyword']
keyword_target = total_df['keyword']
```

```python
from nltk.corpus import stopwords
nlist = []

rx = re.compile(r'(?<=[a-z])(?=[A-Z])')
cnt = 0

for i in tlist:
    cln = i.replace('\n',' ')
    cln = cln.replace('•','')
    cln = rx.sub(' ',cln)
    cln = ' '.join([word for word in cln.split() if word not in stopwords.words('english')])
    cln =''.join( [w for w in cln if w not in string.punctuation])
    cln = cln.lower()
    nlist.append (cln)
    cnt += 1
    if cnt % 100 == 0 : print(cnt,cln[:200])
```

```python
import joblib

joblib.dump(nlist,'nlist.pkl')
```

```python
X = nlist
y = keyword_target
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.30, random_state=43)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
vectorizer = CountVectorizer(stop_words={'english'},min_df=20, ngram_range=(1,2))

cv = vectorizer.fit_transform(nlist)
cv.shape
#vectorizer.get_feature_names()
```

```python
tf_transformer = TfidfTransformer(use_idf=True).fit(cv)
cv_tfidf = tf_transformer.transform(cv)
cv_tfidf.shape
#cv_tfidf.indices
#cv_tfidf.maxprint
#cv_tfidf.minimum
#print(cv_tfidf.sorted_indices())
```

```python
## Training a classifier
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(cv_tfidf, keyword_target)
```

```python
clf.class_count_
clf.classes_
clf.get_params()
clf.score(cv_tfidf, keyword_target)
clf.feature_count_
clf.coef_
clf.predict(cv_tfidf)
```

```python
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

```python
print(classification_report(keyword_target,clf.predict(cv_tfidf)))
```

```python
print(confusion_matrix(keyword_target,clf.predict(cv_tfidf)))
```
